```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
portatiles = pd.read_csv('train.csv')
```

```python
portatiles.info()
```

```python
portatiles.head()
```

```python
portatiles['Ram'] = portatiles['Ram'].str.replace('GB','').astype(int)
portatiles['Weight'] = portatiles['Weight'].str.replace('kg','').astype(float)
```

```python
variables = ['Inches', 'Ram', 'Weight']
```

```python
X = portatiles[variables]
y = portatiles['Price_euros']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)
```

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
modelo.fit(X_train, y_train)

print(f'MAE: {mean_absolute_error(y_test, modelo.predict(X_test))}')
```

```python
from sklearn.tree import DecisionTreeRegressor

modelo = DecisionTreeRegressor(random_state=11)
modelo.fit(X_train, y_train)

print(f'MAE: {mean_absolute_error(y_test, modelo.predict(X_test)):.2f}')
```

```python
import pickle

pickle.dump(modelo, open('modelo_v1.pkl', 'wb'))
```
